```python
import httpx
from bs4 import BeautifulSoup
url = "https://www.audiosciencereview.com/forum/index.php?threads/review-apple-vs-google-usb-c-headphone-adapters.5541/"
url = "https://www.apple.com/jobs/pdf/HBR_How_Apple_Is_Organized_For_Innovation-4.pdf"
```

```python
r = httpx.get(url)
print(r.status_code)

soup = BeautifulSoup(r.text, 'lxml')
# print(soup.title.get_text(strip=True))
print(soup.prettify())
```

```python
r.headers.get("Content-Type")
```

```python
"\n".join([p.get_text(" ", strip=True) for p in soup.find_all("p")])
```

```python
async with httpx.AsyncClient() as client:
    response = await client.get(url)
    response.raise_for_status() # TODO: error handling

    soup = BeautifulSoup(response.text, 'lxml')
    print(soup.prettify())
```
