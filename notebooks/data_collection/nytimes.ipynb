```python
import os
from dotenv import load_dotenv
import requests

# load environment variables
load_dotenv()

NYTIMES_API = os.getenv("NYTIMES_API_KEY")
```

# Most Popular
https://developer.nytimes.com/docs/most-popular-product/1/overview

- Most viewed
- Most emailed
- Most shared 
  
in the last 1 or 7 or 30 days.

You can even the find most shared articles on specific platforms in the last n days (e.g. Facebook)

```python
period = 1
params = {
    "api-key": NYTIMES_API
}
r = requests.get(f"https://api.nytimes.com/svc/mostpopular/v2/viewed/{period}.json", params=params)
res = r.json()
res
```

# Top Stories
https://developer.nytimes.com/docs/top-stories-product/1/overview

Top stories by section

```python
section = "business"
params = {
    "api-key": NYTIMES_API
}
r = requests.get(f"https://api.nytimes.com/svc/topstories/v2/{section}.json", params=params)
res = r.json()
res
```

# Article Search
https://developer.nytimes.com/docs/articlesearch-product/1/overview

```python
query = "Microsoft Stocks"
filter = ""
params = {
    "api-key": NYTIMES_API,
    "q": query,
    "filter": filter
}
r = requests.get(f"https://api.nytimes.com/svc/search/v2/articlesearch.json", params=params)
res = r.json()
res
```

# Scrape

```python
url = "https://www.nytimes.com/2024/06/17/technology/-crypto-influence-election.html"
import json
with open("../../.nyt-headers.json", "r") as f:
    headers = json.load(f)
```

```python
import httpx

with httpx.Client(headers=headers) as client:
    response = client.get(url)

print(response.status_code)
print(response.text)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text, 'lxml')
```

```python
print("\n".join([p.text.strip() for p in soup.findAll('p')]))
```
